```
try:
  import base64  
  import requests
  master = "https://github.com/Xethron/Hangman/blob/master/words.txt?raw=true"
  req = requests.get(master)
  req = req.text
  data = req.split('\n')  
  word_list = []
  for x in data:
    list_word = x.lower()
    word_list.append(list_word)
except:
  word_list = ['apple','account','banana','jacket','tornado','moose','strawberry','coconut','python','pineapple']
import random
 
def replay():
  global incGuess
  global correct
  global cWord
  global word
  global word_to_guess
  global wordVisual
  global guessed
  global incorrect
  global Lguess
  retry = ('')
  while not retry.lower() == 'yes' and not retry.lower() == 'no':
    retry = input('Type yes or no: ')
    if retry.lower() == 'yes':
      correct = False
      cWord = []
      word = rWord()
      word_to_guess = []
      wordVisual = []
      guessed = []
      incorrect = 0
      incGuess = []
      Lguess = ('')
      for letter in word:
        word_to_guess.append(letter)
      for y in range(len(word)):
        wordVisual.append('_')
        cWord.append('_')
    elif retry.lower() == 'no':
      correct = True
    else:
      print('Please type either yes or no')
 
def rWord():
  global word_list
  rNumber = random.randint(0,len(word_list)-1)
  word = word_list[rNumber]
  return word.lower()
def gallows(x):
  noose1 = ('    _____')
  noose2 = ('   |     |')
  noose3 = ('         |')
  noose4 = ('         |')
  noose5 = ('         |')
  noose6 = ('         |')
  noose7 = ('___________')
  noosei3 = ('   O     |')
  noosei4a = ('   |     |')
  noosei4b = ('  /|     |')
  noosei4c = ('  /|\    |')
  noosei5a = ('  /      |')
  noosei5b = ('  / \    |')
  if x == 0:
    print(noose1)
    print(noose2)
    print(noose3)
    print(noose4)
    print(noose5)
    print(noose6)
    print(noose7)
  elif x == 1:
    print(noose1)
    print(noose2)
    print(noosei3)
    print(noose4)
    print(noose5)
    print(noose6)
    print(noose7)
  elif x == 2:
    print(noose1)
    print(noose2)
    print(noosei3)
    print(noosei4a)
    print(noose5)
    print(noose6)
    print(noose7)
  elif x == 3:  
    print(noose1)
    print(noose2)
    print(noosei3)
    print(noosei4b)
    print(noose5)
    print(noose6)
    print(noose7)
  elif x == 4:
    print(noose1)
    print(noose2)
    print(noosei3)
    print(noosei4c)
    print(noose5)
    print(noose6)
    print(noose7)
  elif x == 5:
    print(noose1)  
    print(noose2)
    print(noosei3)
    print(noosei4c)
    print(noosei5a)
    print(noose6)
    print(noose7)
  elif x == 6:
    print(noose1)  
    print(noose2)
    print(noosei3)
    print(noosei4c)
    print(noosei5b)
    print(noose6)
    print(noose7)
 
correct = False
wins = 0
loss = 0
cWord = []
incGuess = []
word = rWord()
word_to_guess = []
wordVisual = []
guessed = []
placement = []
incorrect = 0
for letter in word:
  word_to_guess.append(letter)
 
for x in range(len(word)):
  wordVisual.append('_')
  cWord.append('_')
 
while correct == False:
  guess = ('')
  Lguess = ('')
  gallows(incorrect)
  print('These are the letters you have guessed so far')
  for l in guessed:
    if not l in word_to_guess and not l in incGuess:
      incGuess.append(l)
  print(*incGuess)
  print(*wordVisual)
  if cWord == word_to_guess:
    print('You guessed the word, congrats!')    
    wins = wins + 1
    print('You have won ' + str(wins) + ' times so far and lost '+ str(loss) + ' times')
    print('Would you like to play again?')
    replay()
  elif incorrect < 6:
    print('Type a letter to guess or the full word:')
    Lguess = input()
  elif incorrect == 6:
    loss = loss + 1
    print("That's too bad. The correct word was " + word.upper())
    print('You have won ' + str(wins) + ' times so far and lost '+ str(loss) + ' times')
    print('Would you like to play again?')
    replay()
  if Lguess.isalpha():
    guess = Lguess.lower()
  elif not Lguess.isalpha() and correct == False:
    print('Please type in only letters')  
  if guess == word:
    print('Congrats you guessed the word')
    wins = wins + 1
    print('You have won ' + str(wins) + ' times so far and lost '+ str(loss) + ' times')
    print('Would you like to play again?')
    replay()
  elif not guess == word and not correct == True:
    if guess in word_to_guess and not guess in guessed and len(guess) == 1:
      guessed.append(guess)
      for y in range(len(word_to_guess)):
        if word_to_guess[y] == guess:
          placement.append(y)
      for i in placement:
        wordVisual[i] = guess.upper()
        cWord[i] = guess
      placement = []
    elif guess in guessed and len(guess) == 1 and Lguess.isalpha():
      print('You have already guessed that letter')
    elif not guess in word_to_guess and len(guess) == 1 and Lguess.isalpha():
      guessed.append(guess)
      incorrect = incorrect + 1
    elif len(guess) < len(word) and Lguess.isalpha():
      print('Please type single letters of the full word only please')
```
